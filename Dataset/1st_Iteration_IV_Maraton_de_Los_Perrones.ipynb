<h2>Transformation</h2>

```python
# Check for remaining null values in each column
mthonmerged.isnull().sum()
```

```python
# Analyze possible choices to fill 'Lugar de Residencia' column. In case there are duplicate choices of the same, transform them into one
mthonmerged['Lugar de Residencia'].unique()
```

```python
mthonmerged.groupby("Lugar de Residencia").count()
```

```python
# Merge similar fields into one. Also, remove the state of each city and make a separate column for them.
mthonmerged = mthonmerged.replace('Hermosillo, Sonora', 'Hermosillo')
mthonmerged = mthonmerged.replace('Hillo., Son.', 'Hermosillo')
mthonmerged = mthonmerged.replace('Caborca, Sonora', 'Caborca')
mthonmerged = mthonmerged.replace('Cananea, Sonora', 'Cananea')
mthonmerged = mthonmerged.replace('Cd. Obregón, Son', 'Cd. Obregón')
mthonmerged = mthonmerged.replace('Agua Prieta, Sonora', 'Agua Prieta')
mthonmerged = mthonmerged.replace('Cuernavaca, Morelos', 'Cuernavaca')
mthonmerged = mthonmerged.replace('Culiacán, Sinaloa', 'Culiacán')
mthonmerged = mthonmerged.replace('Ensenada, B.C.', 'Ensenada')
mthonmerged = mthonmerged.replace('Guaymas, Sonora', 'Guaymas')
mthonmerged = mthonmerged.replace('León, Gto.', 'León')
mthonmerged = mthonmerged.replace('Mexicali, B.C.', 'Mexicali')
mthonmerged = mthonmerged.replace('Monterrery, N.L.', 'Monterrey')
mthonmerged = mthonmerged.replace('Nogales, Sonora', 'Nogales')
mthonmerged = mthonmerged.replace('Phoenix, Az. USA', 'Phoenix')
mthonmerged = mthonmerged.replace('San Diego, CA. USA', 'San Diego')
mthonmerged = mthonmerged.replace('Tijuana, B.C.', 'Tijuana')
mthonmerged = mthonmerged.replace('Villa Juarez, Sonora', 'Villa Juarez')
mthonmerged = mthonmerged.replace('G', 'L')
```

```python
mthonmerged.groupby("Lugar de Residencia").count()
```

```python
# Title change to better match with the future 'Estado' column
mthonmerged = mthonmerged.rename(columns={'Lugar de Residencia': 'Ciudad'})
mthonmerged
```

<h2>Fill empty values</h2>

```python
mthonmerged.isnull().sum()
```

```python
# Fill missing values with mode
mthonmerged['Ciudad'] = mthonmerged['Ciudad'].fillna(mthonmerged['Ciudad'].mode()[0])
mthonmerged['Talla'] = mthonmerged['Talla'].fillna(mthonmerged['Talla'].mode()[0])

# Fill missing values with mean (rounded, since it is age)
mthonmerged['Edad'] = mthonmerged['Edad'].fillna(round(mthonmerged.Edad.mean()))
```

```python
# Make sure there are no more empty fields
mthonmerged.isnull().sum()
```

```python
# Function to assign a State to each city
def assign_state(City):
    if City in ['Agua Prieta', 'Bacobampo', 'Cananea', 'Cd. Obregón', 'Guaymas', 'Hermosillo', 'Nogales', 'Caborca', 'San Luis Río Colorado', 'Villa Juarez']:
        return 'Sonora'
    elif City in ['Ensenada', 'Mexicali', 'Tijuana']:
        return 'Baja California'
    elif City == 'Phoenix':
        return 'Arizona, USA'
    elif City == 'San Diego':
        return 'California, USA'
    elif City == 'CDMX':
        return 'CDMX'
    elif City == 'Cuernavaca':
        return 'Morelos'
    elif City == 'Culiacán':
        return 'Sinaloa'
    elif City == 'Monterrey':
        return 'Nuelo León'
    elif City == 'León':
        return 'Guanajuato'
    else:
        return 'Otro'
```

```python
# Apply state assignment function
mthonmerged['Estado'] = mthonmerged['Ciudad'].apply(assign_state)
```

```python
# Set figure size
plt.figure(figsize=(12, 6))

# Plot combined count plot for marathon distances by gender
plt.subplot(1, 2, 1)
ax1 = sns.countplot(x='Edad', hue='Maraton', data=mthonmerged, palette='viridis')
plt.title('Número de participantes en el maratón por edades')
plt.xlabel('Grupo de Edades')
plt.ylabel('Número de Participantes')
plt.legend(title='Ruta', labels=['21 km', '42 km'])
plt.xticks(rotation=45)

# Add counts on top of the bars for the first subplot
for p in ax1.patches:
    ax1.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

# Plot combined count plot for gender by age group
plt.subplot(1, 2, 2)
ax2 = sns.countplot(x='Edad', hue='Genero', data=mthonmerged, palette='magma')
plt.title('Número de participantes en el maratón por edades')
plt.xlabel('Grupo de Edades')
plt.ylabel('Número de Participantes')
plt.legend(title='Género', labels=['Femenino', 'Masculino'])

# Add counts on top of the bars for the second subplot
for p in ax2.patches:
    ax2.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

# Adjust layout
plt.tight_layout()

# Show plot
plt.xticks(rotation=45)
plt.show()
```

```python
# Function to show percentage and number value
def func(pct, valores):
    absolute = int(round(pct / 100. * np.sum(valores)))
    return "{:.1f}%\n({:d})".format(pct, absolute)

myexplode1 = [0, 0, 0.05]
plt.figure(figsize=(8,8))

plt.pie(
    x=mthonstaff,
    labels=['21 km', '42 km', 'Staff'],
    # show percentage with two decimal points
    autopct=lambda pct: func(pct, mthonstaff),
    # increase the size of all text elements
    textprops={'fontsize':14},
    colors=sns.color_palette('Set2'),
    startangle = 90,
    explode = myexplode1
)

# Add Title
plt.title(
    label="Participantes del maratón 2024",
    fontdict={"fontsize":16},
    pad=20
)

plt.show()
```

```python
# Calculate number of participants per city
city_counts = mthonmerged['Ciudad'].value_counts()

# Order plot by largest count of participants per city to smallest
city_ordered = city_counts.index.tolist()

# Set figure size
plt.figure(figsize=(12, 8))

# Create the countplot with horizontal bars, utilizando los estados ordenados
ax = sns.countplot(y='Ciudad', data=mthonmerged, palette='viridis', order=city_ordered)

# Add counts on the right side of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_width()}', (p.get_width(), p.get_y() + p.get_height() / 2.),
                ha='center', va='center', fontsize=12, color='black', xytext=(14, 0),
                textcoords='offset points')

plt.title('Número de participantes en el maratón por ciudad')
plt.xlabel('Número de Participantes')
plt.ylabel('Ciudad')

# Adjust layout
plt.tight_layout()

plt.show()
```

```python
# Calculate number of participants per state
state_counts = mthonmerged['Estado'].value_counts()

# Order plot by largest count of participants per state to smallest
state_ordered = state_counts.index.tolist()

# Set figure size
plt.figure(figsize=(12, 8))

# Create the countplot with horizontal bars, utilizando los estados ordenados
ax = sns.countplot(y='Estado', data=mthonmerged, palette='viridis', order=state_ordered)

# Add counts on the right side of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_width()}', (p.get_width(), p.get_y() + p.get_height() / 2.),
                ha='center', va='center', fontsize=12, color='black', xytext=(14, 0),
                textcoords='offset points')

plt.title('Número de participantes en el maratón por estado')
plt.xlabel('Número de Participantes')
plt.ylabel('Estado')

# Adjust layout
plt.tight_layout()

plt.show()
```
