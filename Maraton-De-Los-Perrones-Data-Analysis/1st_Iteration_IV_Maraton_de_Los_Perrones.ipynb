<h2>Transformation</h2>

```python
# Remove empty rows based on 'Nombre' column
mthonmerged = mthonmerged.dropna(subset=['Nombre'])
mthonmerged
```

```python
# Check for remaining null values in each column
mthonmerged.isnull().sum()
```

```python
# Analyze possible choices to fill 'Lugar de Residencia' column. In case there are duplicate choices of the same, transform them into one
mthonmerged['Lugar de Residencia'].unique()
```

```python
mthonmerged.groupby("Lugar de Residencia").count()
```

```python
# Merge similar fields into one. Also, remove the state of each city and make a separate column for them.
mthonmerged = mthonmerged.replace('Hermosillo, Sonora', 'Hermosillo')
mthonmerged = mthonmerged.replace('Hillo., Son.', 'Hermosillo')
mthonmerged = mthonmerged.replace('Caborca, Sonora', 'Caborca')
mthonmerged = mthonmerged.replace('Cananea, Sonora', 'Cananea')
mthonmerged = mthonmerged.replace('Cd. Obreg√≥n, Son', 'Cd. Obreg√≥n')
mthonmerged = mthonmerged.replace('Agua Prieta, Sonora', 'Agua Prieta')
mthonmerged = mthonmerged.replace('Cuernavaca, Morelos', 'Cuernavaca')
mthonmerged = mthonmerged.replace('Culiac√°n, Sinaloa', 'Culiac√°n')
mthonmerged = mthonmerged.replace('Ensenada, B.C.', 'Ensenada')
mthonmerged = mthonmerged.replace('Guaymas, Sonora', 'Guaymas')
mthonmerged = mthonmerged.replace('Le√≥n, Gto.', 'Le√≥n')
mthonmerged = mthonmerged.replace('Mexicali, B.C.', 'Mexicali')
mthonmerged = mthonmerged.replace('Monterrery, N.L.', 'Monterrey')
mthonmerged = mthonmerged.replace('Nogales, Sonora', 'Nogales')
mthonmerged = mthonmerged.replace('Phoenix, Az. USA', 'Phoenix')
mthonmerged = mthonmerged.replace('San Diego, CA. USA', 'San Diego')
mthonmerged = mthonmerged.replace('Tijuana, B.C.', 'Tijuana')
mthonmerged = mthonmerged.replace('Villa Juarez, Sonora', 'Villa Juarez')
mthonmerged = mthonmerged.replace('G', 'L')
```

```python
mthonmerged.groupby("Lugar de Residencia").count()
```

```python
# Title change to better match with the future 'Estado' column
mthonmerged = mthonmerged.rename(columns={'Lugar de Residencia': 'Ciudad'})
mthonmerged
```

```python
# Checkpoint to add future missing values
save_checkpoint = 'Datos_Limpios_Incompletos_Inscripciones_Maratones.xlsx'
mthonmerged.to_excel(save_checkpoint, index = False)
```

<h2>Fill empty values</h2>

```python
mthonmerged.isnull().sum()
```

```python
# Calculate the mode of the column 'Ciudad'
moda_ciudad = mthonmerged['Ciudad'].mode()
moda_ciudad[0]
```

```python
# Fill empty fields with the mode
mthonmerged['Ciudad'].fillna(moda_ciudad[0], inplace = True)
```

```python
# Same proccess with column 'Talla'
moda_talla = mthonmerged['Talla'].mode()
moda_talla[0]
```

```python
mthonmerged['Talla'].fillna(moda_talla[0], inplace = True)
```

```python
# For 'Edad' column, do mean instead of mode
mean_edad = round(mthonmerged['Edad'].mean())
mthonmerged['Edad'].fillna(mean_edad, inplace = True)
```

```python
# Make sure there are no more empty fields
mthonmerged.isnull().sum()
```

```python
# Function to assign a State to each city
def assign_state(City):
    if City in ['Agua Prieta', 'Bacobampo', 'Cananea', 'Cd. Obreg√≥n', 'Guaymas', 'Hermosillo', 'Nogales', 'Caborca', 'San Luis R√≠o Colorado', 'Villa Juarez']:
        return 'Sonora'
    elif City in ['Ensenada', 'Mexicali', 'Tijuana']:
        return 'Baja California'
    elif City == 'Phoenix':
        return 'Arizona, USA'
    elif City == 'San Diego':
        return 'California, USA'
    elif City == 'CDMX':
        return 'CDMX'
    elif City == 'Cuernavaca':
        return 'Morelos'
    elif City == 'Culiac√°n':
        return 'Sinaloa'
    elif City == 'Monterrey':
        return 'Nuelo Le√≥n'
    elif City == 'Le√≥n':
        return 'Guanajuato'
    else:
        return 'Otro'
```

```python
# Apply state assignment function
mthonmerged['Estado'] = mthonmerged['Ciudad'].apply(assign_state)
```

```python
# Function to show percentage and number value
def func(pct, valores):
    absolute = int(round(pct / 100. * np.sum(valores)))
    return "{:.1f}%\n({:d})".format(pct, absolute)

myexplode1 = [0, 0, 0.05]
plt.figure(figsize=(8,8))

plt.pie(
    x=mthonstaff,
    labels=['21 km', '42 km', 'Staff'],
    # show percentage with two decimal points
    autopct=lambda pct: func(pct, mthonstaff),
    # increase the size of all text elements
    textprops={'fontsize':14},
    colors=sns.color_palette('Set2'),
    startangle = 90,
    explode = myexplode1
)

# Add Title
plt.title(
    label="Participantes del marat√≥n 2024",
    fontdict={"fontsize":16},
    pad=20
)

plt.show()
```

```python
# Calculate number of participants per city
city_counts = mthonmerged['Ciudad'].value_counts()

# Order plot by largest count of participants per city to smallest
city_ordered = city_counts.index.tolist()

# Set figure size
plt.figure(figsize=(12, 8))

# Create the countplot with horizontal bars, utilizando los estados ordenados
ax = sns.countplot(y='Ciudad', data=mthonmerged, palette='viridis', order=city_ordered)

# Add counts on the right side of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_width()}', (p.get_width(), p.get_y() + p.get_height() / 2.),
                ha='center', va='center', fontsize=12, color='black', xytext=(14, 0),
                textcoords='offset points')

plt.title('N√∫mero de participantes en el marat√≥n por ciudad')
plt.xlabel('N√∫mero de Participantes')
plt.ylabel('Ciudad')

# Adjust layout
plt.tight_layout()

plt.show()
```
